```python
def clean_data(doc):
    metin = doc.replace("\n", " ").replace("\r", " ").replace("\t", " ")
    while "  " in metin:
        metin = metin.replace("  ", " ")
    return metin.strip()
```

```python
# Define a function that scrapes the questions and answers from a given page

def scrape_all():
    """
    This function scrapes all the questions and answers from the website 
    and returns them as a pandas DataFrame.

    Returns:
        df: pandas DataFrame
    """
    all_question_answers = []
    for i in range(1, 37):
        response = requests.get("https://www.hukuksorucevap.com.tr/sorucevap/?sayfa=" + str(i))
        response.encoding = "utf-8"
        soup = BeautifulSoup(response.text(), "html.parser")
        question = soup.find("div", {"id": "Soru"})
        questions = question.find_all("li")
        for question_li in questions:
            questions_dic = {}
            question_text = question_li.find('div', {'id': 'Soru'}).text
            questions_dic["soru"] = clean_data(question_text)
            answer_text = question_li.find('div', {'id': 'Cevap'}).find("div", {"class": "DoktorCevabi"}).text
            questions_dic["cevap"] = clean_data(answer_text)
            statu_of_replier = question_li.find('div', {'id': 'Cevap'}).find("div", {"class": "DoktorUnvani"}).text
            questions_dic["cevaplayan_unvan"] = clean_data(statu_of_replier)
            all_question_answers.append(questions_dic)
    return pd.DataFrame(all_question_answers)

df = scrape_all()

           

```

# Paralel Scarping (Async)

```python
import time
import asyncio
import aiohttp
import json
```

```python
all_question_answers = []
async def parallel_scrape_all(session,i):
    """
    This async function scrapes all the questions and answers from the website
    """
    try:
        async with session.get("https://www.hukuksorucevap.com.tr/sorucevap/?sayfa=" + str(i)) as response:
            response.encoding = "utf-8"
            soup = BeautifulSoup(await response.text(), "html.parser")
            question = soup.find("div", {"id": "Soru"})
            questions = question.find_all("li")
            for question_li in questions:
                questions_dic = {}
                question_text = question_li.find('div', {'id': 'Soru'}).text
                questions_dic["soru"] = clean_data(question_text)
                answer_text = question_li.find('div', {'id': 'Cevap'}).find("div", {"class": "DoktorCevabi"}).text
                questions_dic["cevap"] = clean_data(answer_text)
                statu_of_replier = question_li.find('div', {'id': 'Cevap'}).find("div", {"class": "DoktorUnvani"}).text
                questions_dic["cevaplayan_unvan"] = clean_data(statu_of_replier)
                all_question_answers.append(questions_dic)
    except Exception as e:
        print(e)
```

```python
start = time.time()
async with aiohttp.ClientSession() as session:
    tasks = []
    for i in range(1, 37):
        tasks.append(parallel_scrape_all(session, i))
    await asyncio.gather(*tasks)
end = time.time()
print(f"Time taken to get the details of profiles: {end - start} seconds")
```
